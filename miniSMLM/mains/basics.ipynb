```python
import numpy as np
import matplotlib.pyplot as plt
from miniSMLM import generators, localize, utils, psf
```

### Parameters 

```python
config = {
    "sigma": 0.92,
    "N0": 1000,
    "B0": 0,
    "eta": 1.0,
    "texp": 1.0,
    "gain": 1.0,
    "offset": 100.0,
    "var": 5.0
}
```

### Generate some toy data of emitters spaced on a ring

```python
nparticles=3;radius=10.0;npixels=50
disc2d = generators.Disc2D(npixels,npixels)
adu,_,_ = disc2d.forward(radius,nparticles,**config,show=True)
```

### Run a detection algorithm to localize the emitters

```python
print(adu.dtype)
detector = localize.LoGDetector(adu,threshold=10.0)
spots = detector.detect()
detector.show(adu)
```

### Run a maximum likelihood estimation to localize the emitters with sub-pixel precision

```python
cam_params = [config['eta'],config['texp'],config['gain'],
                           config['offset'],config['var']]
patchw=3; max_iters=1000
for i in spots.index:
    x0 = int(spots.at[i,'x'])
    y0 = int(spots.at[i,'y'])
    patch = adu[x0-patchw:x0+patchw+1,y0-patchw:y0+patchw+1]
    patch = patch - cam_params[3]
    patch = np.clip(patch,0,None)
    theta0 = np.array([patchw,patchw,config['N0']])
    opt = psf.MLE2D_BFGS(theta0,patch,config) #cartesian coordinates with top-left origin
    theta_mle, loglike, converged, err = opt.optimize(max_iters=max_iters,plot_fit=True)
    dx = theta_mle[1] - patchw; dy = theta_mle[0] - patchw
    spots.loc[i, 'x_mle'] = x0 + dx #switch back to image coordinates
    spots.loc[i, 'y_mle'] = y0 + dy
    spots.loc[i, 'N0'] = theta_mle[2]
```

### Generate the super-resolution image by Gaussian kernel density estimation (KDE)

```python
kde = utils.KDE(spots)
render = kde.forward()
fig,ax=plt.subplots()
ax.imshow(render,cmap='gray'); plt.show()
```
