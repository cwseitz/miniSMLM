```python
from DNAFlex.psf import GaussianPSF, hessian_analytical, hessian_autograd
from scipy.special import erf
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 19
omat = np.ones((L,L))
gain0 = 2.2
rmu0 = 0
rvar0 = 0
gain = gain0*omat #ADU/e-
rmu = rmu0*omat #ADU
rvar = rvar0*omat #ADU^2
pixel_size = 108.3 #nm
sigma = 0.22*640/1.4 #zhang 2007
sigma = sigma = sigma/pixel_size
N0 = 5000 #cps
texp = 0.1 #seconds
eta = 0.5 #quantum efficiency
I0 = eta*N0*texp #expected number of photons during exposure time (assume lam0 is determined for now)
x0,y0 = (8.0,5.0)

print(f'I0={I0}')
psf = GaussianPSF(x0,y0,L,I0,sigma,gain,rmu,rvar)
adu = psf.generate(plot=True)
counts = np.round(adu/gain).astype(int)
```

```python
#Define parameter guess
x0,y0,sigma,N0 = 8.0,5.0,1.0,5000.0 #parameter guess
theta = np.array([x0,y0,sigma,N0])
```

```python
H1 = hessian_autograd(theta, counts, eta, texp)
```

```python
H2 = hessian_analytical(theta, counts, eta, texp)
```

```python
fig, ax = plt.subplots(1,2)
ax[0].imshow(H1)
ax[1].imshow(H2)
plt.show()
```
