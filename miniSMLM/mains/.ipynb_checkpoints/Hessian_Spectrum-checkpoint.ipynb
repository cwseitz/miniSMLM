```python
from DNAFlex.psf import *
from scipy.stats import multivariate_normal
import numpy as np
```

```python
#########################
# Parameters
#########################

L = 15
omat = np.ones((L,L))
gain0 = 2.2 #huang 2013
rmu0 = 0
rvar0 = 0
gain = gain0*omat #ADU/e-
rmu = rmu0*omat #ADU
rvar = rvar0*omat #ADU^2
pixel_size = 108.3 #nm
sigma = 0.22*640/1.4 #zhang 2007
sigma = sigma = sigma/pixel_size
texp = 0.01 #seconds
eta = 0.8 #quantum efficiency
x0,y0 = (8.0,8.0)
```

```python
################################################
# Show some example data in different conditions
################################################

fig, ax = plt.subplots(2,3)

n0, b0 = 100000,2000 #huang 2013
psf = GaussianPSF(x0,y0,n0,b0,sigma,eta,texp,L,gain,rmu,rvar)
adu = psf.generate(plot=False)
ax[0,0].imshow(adu)
n0, b0 = 100000,2000
psf = GaussianPSF(x0,y0,n0,b0,sigma,eta,texp,L,gain,rmu,rvar)
adu = psf.generate(plot=False)
ax[0,1].imshow(adu)
n0, b0 = 100000,2000
psf = GaussianPSF(x0,y0,n0,b0,sigma,eta,texp,L,gain,rmu,rvar)
adu = psf.generate(plot=False)
ax[0,2].imshow(adu)
n0, b0 = 100000,2000
psf = GaussianPSF(x0,y0,n0,b0,sigma,eta,texp,L,gain,rmu,rvar)
adu = psf.generate(plot=False)
ax[1,0].imshow(adu)
n0, b0 = 100000,2000
psf = GaussianPSF(x0,y0,n0,b0,sigma,eta,texp,L,gain,rmu,rvar)
adu = psf.generate(plot=False)
ax[1,1].imshow(adu)
n0, b0 = 100000,2000
psf = GaussianPSF(x0,y0,n0,b0,sigma,eta,texp,L,gain,rmu,rvar)
adu = psf.generate(plot=False)
ax[1,2].imshow(adu)
plt.show()
```

```python
########################################
# Check the error as a function of N0/B0
########################################

Nspace = np.linspace(500/texp,2000/texp,100)
Bspace = np.array([1,5,20])/texp
err_arr = np.zeros((5,100,3)) #param,n0,b0

for i, b0 in enumerate(Bspace):
    for j, n0 in enumerate(Nspace):
        psf = GaussianPSF(x0,y0,n0,b0,sigma,eta,texp,L,gain,rmu,rvar)
        adu = psf.generate(plot=False)
        counts = np.round(adu/gain).astype(int)
        theta = np.array([x0,y0,sigma,n0,b0])
        H = hessian_analytical(theta,counts,eta,texp)
        Hinv = np.linalg.inv(H)
        err = np.sqrt(np.diag(Hinv))
        err_arr[:,j,i] = err
```

```python
fig, ax = plt.subplots(1,5,figsize=(12,3))
colors = ['red','blue','purple']
for n in range(len(Bspace)):
    ax[0].plot(Nspace,err_arr[0,:,n]*pixel_size,label=f'bg={Bspace[n]}',color=colors[n])
    ax[1].plot(Nspace,err_arr[1,:,n]*pixel_size,color=colors[n])
    ax[2].plot(Nspace,err_arr[2,:,n]*pixel_size,color=colors[n])
    ax[3].plot(Nspace,err_arr[3,:,n],color=colors[n])
    ax[4].plot(Nspace,err_arr[4,:,n],color=colors[n])
for axi in ax:
    axi.set_xlabel(r'$N_{0}$')
ax[0].set_ylabel(r'$\Delta x$ (nm)')
ax[1].set_ylabel(r'$\Delta y$ (nm)')
ax[2].set_ylabel(r'$\Delta \sigma$ (nm)')
ax[3].set_ylabel(r'$\Delta N_{0}$ (photons)')
ax[4].set_ylabel(r'$\Delta B_{0}$ (photons)')
plt.tight_layout()
ax[0].legend()
plt.show()
```
